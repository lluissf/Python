```python
notas = []
baixas = []
soma = 0
for i in range(30):
    notas.append(float(input(f"Digite a {i+1}° nota: ")))
    soma += notas[i]
# Imprimindo a maior e a menor nota:
notas.sort()
menor_nota = notas[0]
notas.reverse()
maior_nota = notas[0]
# Calculando a média e definindo a quantidade de nota baixa.
media = soma / 30.0
for nota in notas:
    if nota < media:
        baixas.append(nota)


print(f"A Maior nota é {maior_nota}\n A menor nota é {menor_nota} \n A média da turma é: {media} \n Possui: {len(baixas)}  notas baixas.")
```

```python
mercadorias = list(range(3)) # 100
faturamento = 0
preco = []
quantidade = []

for i in mercadorias:
  quantidade.append(int(input(f"Qual a quantidade da {i+1}° mercadoria que foram vendidas neste mês ?  ")))
  preco.append(float(input(f"Qual o valor da {i+1} mercadoria ? ")))
print("======================================")
print("Mercadoria \t Quantidade \t Preço")
for i in mercadorias:
  faturamento += quantidade[i] * preco[i]
  print("======================================")
  print(f"{i+1} \t\t {quantidade[i]} \t\t R${preco[i]} ")
print("======================================")
print(f"O Faturamento Mensal é R${faturamento: .2f}")
```
