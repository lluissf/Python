```python
def fahrenheit(F):
    C = (F-32) * 5/9
    return f"O valor convertido é de {C:.1f}°Celsius"

f = float(input("Forneça a temperatura em Fahrenheit: "))
print(fahrenheit(f))
```

```python
def hipotenusa(catetoA, catetoB):
    hipotenusa = (catetoA**2 + catetoB**2) ** 0.5
    return hipotenusa
valor1 = int(input("Forneça um valor: "))
valor2 = int(input("Forneça outro valor: "))
resultado = hipotenusa(valor1, valor2)
print(f"{resultado: .2f}")
```

```python
def pesoideal(h, sexo):
    """Aqui iremos verificar se é homem ou mulher!"""
    if sexo != 1 and sexo != 2:
        return "Sexo inválido!"
    """Agora iremos calcular a altura!"""
    if sexo == 1:
        peso = (62.1 * h) - 44.7 # Caso for mulher
        sexo = "a Mulher"

    elif sexo == 2:
        peso = (72.7 * h) - 58 # Caso for homem
        sexo = "e Homem"
    return f"O Peso Ideal para ess{sexo} seria de: {peso: .2f}kg"
altura = float(input("Forneça a altura: "))
sexo = int(input("Qual o sexo? Feminino (1) ou Masculino (2): "))
pesoideal(altura, sexo)
```

```python
def poligono(lados, cm):
    conta = 0

    if lados == 3:
        tipo = "Triângulo"
        conta = lados * cm

    elif lados == 4:
        tipo = "Quadrado"
        conta += lados * lados

    elif lados == 5:
        tipo = "Pentágono"

    else:
        tipo = "Valor de lados inválido!"


    if tipo == "Triângulo":
        return f"{tipo}, o valor do seu perímetro é de {conta}"
    elif tipo == "Quadrado":
        return f"{tipo}, o valor da sua área é de {conta}"
    else:
        return f"{tipo}"

lados = int(input("Quantos lados tem o polígono? (3-5) "))
cm = int(input("Qual a médida de cada lado? (cm) "))
poligono(lados, cm)
```

```python
def soma(n1, n2):
    soma = 0
    for i in range(n1, n2):
        soma += i
    return f"O Resultado da soma entre {n1} e {n2} é: {soma-n1}"
n1 = int(input("Forneça o 1° número: "))
n2 = int(input("Forneça o 2° número: "))
soma(n1, n2)
```

```python
meses = [ "janeiro", "fevereiro", "março", "abril", "maio", "junho", "julho", "agosto", "setembro", "outubro", "novembro", "dezembro"]
def mes(x):
    if x < 1 or x > 12:
        return "O número escolhido não está entre 1 e 12!"
    else:
        return f"O Mês que você escolheu foi {meses[x-1]}"

n = int(input("Forneça o mês (1-12): "))
mes(n)
```

```python
dias = ["domingo", "segunda", "terça", "quarta", "quinta", "sexta", "sabado"]

def dia(x):
    if x < 1 or x > 7:
        return "O número escolhido não está entre 1 e 7!"
    else:
        escolhido = dias[x-1]
        return str.upper(escolhido[0:3])

n = int(input("Forneça um dia da semana (1-7): "))
dia(n)
```

```python
def calcular(n1, n2):
    if n1 > n2:
        return f"{n1} é maior que {n2}"
    elif n1 < n2:
        return f"{n2} é maior que {n1}"
    else:
        return f"Ambos tem o mesmo valor"

n1 = float(input("Insira um valor: "))
n2 = float(input("Insira um valor: "))
calcular(n1, n2)
```
